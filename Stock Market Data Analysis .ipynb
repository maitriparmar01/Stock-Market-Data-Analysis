### Stock Market Data Analysis (Using Libraries)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

#
data = pd.read_csv("stock_data.csv")

print("First 5 rows:")
print(data.head())

# Numpy
avg_price = np.mean(data['Close'])
max_price = np.max(data['Close'])
min_price = np.min(data['Close'])

print("\nAverage Closing Price:", avg_price)
print("Maximum Closing Price:", max_price)
print("Minimum Closing Price:", min_price)

# Highest trading volume
max_volume = data['Volume'].max()
print("Highest Trading Volume:", max_volume)

# Line chart (Matplotlib) 
plt.plot(data['Close'], color='green')
plt.title("Stock Closing Price Trend")
plt.xlabel("Days")
plt.ylabel("Close Price")
plt.show()

# Bar chart 
plt.bar(data['Date'], data['Volume'], color='orange')
plt.title("Trading Volume")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()

# Pie chart 
labels = ['Open Price','Close Price']
values = [data['Open'].mean(), data['Close'].mean()]

plt.pie(values, labels=labels, autopct='%1.1f%%',
        colors=['skyblue','lightgreen'])
plt.title("Average Price Comparison")
plt.show()

# Seaborn
sns.barplot(x=data['Date'], y=data['Close'], color='purple')
plt.title("Close Price by Date")
plt.show()

# Plotly
px.line(data, x='Date', y='Close',
              title="Stock Price Trend (Plotly)").show()
```
